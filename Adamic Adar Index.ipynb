```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as exp
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import random
import math
```

```python
df = pd.read_csv('/content/adamic3.csv')
```

```python
df
```

```python
mean=df['median_house_value'].mean(axis=0)
range=mean/4
```

```python
range
```

```python
df['Adamic_Adar_index']=0
```

```python
indexs=[]
```

```python
for i in df['median_house_value']:
  first=1
  second=1
  third=1
  for j in df['median_house_value']:
    if(abs(j-i)<=range):
      first=first+1
    if(abs(j-i)<=2*range):
      second=second+1
    if(abs(j-i)<=3*range):
      third=third+1
  index=0
  if(first!=1):
    index=(1/math.log(first))
  if(second!=1):
    index=index+(1/math.log(second))
  if(third!=1):
    index=index+(1/math.log(third))
  indexs.append(index)
```

```python
indexs
```

```python
df['Adamic_Adar_index']=indexs
```

```python
df.to_csv('adamic.csv')
```
